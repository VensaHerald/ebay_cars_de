As mentioned above the price and odometer columns require some work as they are currently stored as text.  
The price column contains a dollar sign and commas separating the thousands.  I shall remove these and then change the data type to in64.
The odometer data contains km units and similar commas separating thousands.

After these transformations I will rename the columns to contain units.

```python
autos['price'] = (autos['price']
                  .str.replace("$","")
                  .str.replace(",","")
                  .astype("int64"))

autos['odometer'] = (autos['odometer']
                  .str.replace("km","")
                  .str.replace(",","")
                  .astype("int64"))
```

Looking at the registration_year data there would appear to be some anomolous figures given that the max is the year 9999 and the min is 1000.  The data set also appears to be right skewed. 

As the first cars were registered in the early decades of the 20th century it would be prudent to assume that any listings with the car registered prior to this date can be dropped.  Similarly this survey was carried out in 2016, and so any years of registration after this can also safely be dropped.

```python
year_drops = autos.loc[~autos["registration_year"].between(1900,2016),:]
autos = autos.drop(year_drops.index)
autos["registration_year"].value_counts(normalize=True).sort_index()
```

After removing the outliers the data now appears as expected.  The listings are heavily biased towards recent registration years.  Roughly the last 20 years back to 1996.

I will now investigate the various brands in this dataset to look into which brand of car sold for the most on eBay over this sample period.

I will do this by aggregating the data for the brands of car with greater than 5% of the listings in this data.  This gives the following brands: VW, BMW, Opel, Mercedes, Audi, Ford, Renault.  These are the most common car brands are so are the most relevant to look into the prices of.

```python
mask_5per = autos["brand"].value_counts(normalize=True)>=.04
values_agg = mask_5per[mask_5per==True].index
price_per_brand = {}

for brand in values_agg:
    mean_price = np.floor(autos.loc[autos["brand"]==brand,"price_usd"].mean())
    price_per_brand[brand]=mean_price
    
price_per_brand

```

As shown above the mean prices for the selected brands of cars are ranked as follows.

Audi
Mercedes Benz
BMW
VW
Ford
Opel
Renault

As these are used car sales this could indicate that these brands of cars hold their value better.  However it could also indicate that these cars are also "sold newer" than other brands and so their condition at sale is better.
